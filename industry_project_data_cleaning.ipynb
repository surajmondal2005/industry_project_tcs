```python
import pandas as pd
from sklearn.cluster import KMeans

# =========================
# 1. LOAD DATA
# =========================
df = pd.read_csv("Retail_Sales_Dataset.csv")

# =========================
# 2. FIX COLUMN NAMES (VERY IMPORTANT)
# =========================
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("Columns:", df.columns)

# =========================
# 3. DATA CLEANING
# =========================
df.drop_duplicates(inplace=True)
df.ffill(inplace=True)

# =========================
# 4. DATE FIX
# =========================
df['date'] = pd.to_datetime(df['date'])

# =========================
# 5. FEATURE ENGINEERING
# =========================
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day_name()
df['quarter'] = df['date'].dt.quarter

# Age group
bins = [18,25,35,45,60,100]
labels = ['18-25','26-35','36-45','46-60','60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Total Amount (NOW WILL WORK)
df['total_amount'] = df['quantity'] * df['price_per_unit']

# =========================
# 6. MACHINE LEARNING
# =========================
X = df[['quantity','total_amount']]

kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

# =========================
# 7. SAVE FINAL DATASET
# =========================
df.to_csv("final_sales_data.csv", index=False)

print("\n✅ DONE — FILE SAVED: final_sales_data.csv")
```
